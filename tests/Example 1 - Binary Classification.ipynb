```python
from Algometrix.algometrix import algometrix

import warnings
warnings.filterwarnings("ignore")
```

```python
# binary classification

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

data = pd.read_csv("/content/synthetic_binary.csv")


X = data[["Feature 1", "Feature 2"]]
y = data["Class"]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

result = algometrix(
    X_train,
    X_test,
    y_train,
    y_test,
    prob_type="class",
    classification_type="binary"
)
```
